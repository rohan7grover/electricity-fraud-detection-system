```python
from pycaret.datasets import get_data
```

```python
import pandas as pd
dataset = pd.read_csv('/content/drive/MyDrive/Labelled_Data.csv')
```

```python
dataset = dataset.fillna(0)
```

```python
# check the shape of data
dataset.shape
```

```python
data = dataset.sample(frac=0.70, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.anomaly import *

exp_ano101 = setup(data, normalize = False, preprocess = False,
                   ignore_features = ['Customer_Number'],
                   session_id = 123)
```

```python
iforest = create_model('iforest')
print(iforest)
```

```python
models()
```

```python
iforest_results = assign_model(iforest)
iforest_results.head()
```

```python
unseen_predictions = predict_model(iforest, data=data_unseen)
unseen_predictions.head()
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix

# Assuming you have a DataFrame df with 'Fraud_Status' and 'Anomaly' columns

# Extract the ground truth and predicted labels
y_true = unseen_predictions['Fraud_Status']
y_pred = unseen_predictions['Anomaly']

# Calculate accuracy
accuracy = accuracy_score(y_true, y_pred)
print(f'Accuracy: {accuracy}')

# Create a confusion matrix
confusion = confusion_matrix(y_true, y_pred)
print('Confusion Matrix:')
print(confusion)
```
